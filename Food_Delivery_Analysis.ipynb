```python
import pandas as pd
import json
import re
orders = pd.read_csv('orders.csv')
users = pd.read_json('users.json')

with open('restaurants.sql', 'r') as f:
    sql_content = f.read()
pattern = r"INSERT INTO restaurants VALUES \((.*)\);"
matches = re.findall(pattern, sql_content)
res_data = [eval(f"({m})") for m in matches]
restaurants = pd.DataFrame(res_data, columns=['restaurant_id', 'restaurant_name_sql', 'cuisine', 'rating'])
df = orders.merge(users, on='user_id', how='left')
df = df.merge(restaurants[['restaurant_id', 'cuisine', 'rating']], on='restaurant_id', how='left')
df.to_csv('final_food_delivery_dataset.csv', index=False)
print("--- PREVIEW OF FINAL DATA ---")
display(df.head())

print("\n--- QUICK ANSWERS ---")
print(f"Total Orders: {len(df)}")
print(f"Total Revenue: {df['total_amount'].sum():.2f}")
print(f"Top City (Gold Members): {df[df['membership'] == 'Gold'].groupby('city')['total_amount'].sum().idxmax()}")
```
